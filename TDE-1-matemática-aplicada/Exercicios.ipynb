```python
# exercício 1

# importei a function date da lib datetime para pegar o ano atual e calcular 
# a idade aproximada do usuário, o ano não é exato pois não sei o mês e dia de 
# nascimento, mas é uma boa aproximação.
from datetime import date

# le os inputs e armazena nas variáveis
nome = input("insira o nome: ")
ano_nascimento = int(input("insira o ano de nascimento: "))

# calcula e exibe o resultado, usa date.today().year para pegar o ano atual com base no sistema operacional
print(f"Nome: {nome} | Idade Aproximada: {date.today().year - ano_nascimento} anos")
```

```python
# exercício 12
import sys

s = input("insira uma string de pelo menos 6 caracteres: ")

if len(s) < 6:
    print("a string deve conter pelo menos 6 caracteres")
    sys.exit(1)

print(f"primeiro caractere: {s[0]}\n3 primeiros caracteres: {s[:3]}\n3 ultimos caracteres: {s[-3:]}\ntexto sem o primeiro caractere: {s[1:]}\ntexto sem o ultimo caractere: {s[:-1]}")
```

```python
# exercício 13

# é importante lembrar que em python, a indexação de strings começa em 0
# as posições foram consideradas com base no índice e não nas letras da palavra
s = input("insira uma palavra: ")

# fiz uma lista fazia e fui fazendo append dos caracteres nas posições par e impar, depois exibi as listas
listaPar = []
listaImpar = []

for i in range(len(s)):
    if i % 2 == 0:
        listaPar.append(s[i])
    else:
        listaImpar.append(s[i])

print(f"posição par: {listaPar}\nposição impar: {listaImpar}")
```

```python
# exercício 14
import sys

# valida se os valores são reais e positivos
v = float(input("insira um valor real v (km/h): "))
t = float(input("insira um valor real t (h): "))
if v < 0 or t < 0:
    print("v e t devem ser valores reais positivos")
    sys.exit(1)

# calcula e exite distancia em km e m
distancia = v * t
print(f"distancia em km: {distancia} km\ndistancia em metros: {distancia * 1000} m")
```
