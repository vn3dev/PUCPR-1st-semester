```python
def f(x):
    return x ** 2

def g(p, q):
    s = p**(0.5) - 3 * q
    return s

print(f(3))
print(g(9, 2))
```

```python
import matplotlib.pyplot as plt

def f(x):
    return x**2

x = list(range(-8, 9))
y = [f(i) for i in x]

plt.plot(x, y)
plt.title("f(x) = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [i for i in range(-5, 6)]
y = [i**2 for i in x]

plt.plot(x, y, label="plot (linha)")
plt.scatter(x, y, label="scatter (pontos)")

plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

def f(x): return x**2
def g(x): return 2 * x + 1

x = list(range(-10, 11))

plt.plot(x, [f(i) for i in x])
plt.plot(x, [g(i) for i in x])
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

def f(x):
    return (x**3) - (2 * x) + 1

x = list(range(-10, 10))

plt.plot(x, [f(i) for i in x])
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

def g(x):
    return ((x**2) - (4*x) + 3)/2

x = list(range(-9, 9))

plt.axhline(0)
plt.axvline(0)
plt.plot(x, [g(i) for i in x], label='plot (linha)')
plt.scatter(x, [g(i) for i in x], color='orange', label='scatter (pontos)')
plt.grid()
plt.legend()
plt.show()
```
