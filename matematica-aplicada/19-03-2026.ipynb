```python
valores = [True, False]

print("| p |", "| q |", "| p and q |")
for p in valores:
    for q in valores:
        print(p, q, p and q)
```

```python
valores = [True, False]

print(f"{'p': ^6} {'q': ^6} {'p and q': ^6}")
print("-"*22)

for p in valores:
    for q in valores:
        print(f"{str(p): ^6} {str(q): ^6} {str(p and q): ^6}")
```
