```python
valores = [True, False]

for p in valores:
    for q in valores:
        implicacao = (not p) or q
        print(f"{p} -> {q} = {implicacao}")
```

```python
valores = [True, False]

for p in valores:
    for q in valores:
        print(p, q, p == q)
```

```python
valores = [True, False]

print(f"{'p':^5} {'q':^5} {'p and q':^10} {'not (p and q)':^15}")
for p in valores:
    for q in valores:
        print(f"{(p):^5} {str(q):^5} {str(p and q):^10} {str(not (p and q)):^15}")
```

```python
valores = [True, False]

print(f"{'p':^5} {'q':^5} {'p or q':^10} {'(p or q) and (not p)':^15}")
for p in valores:
    for q in valores:
        print(f"{str(p):^5} {str(q):^5} {str(p or q):^10} {str((p or q) and (not p)):^15}")
```

```python
valores = [True, False]

print(f"{'p':^5} {'q':^5} {'not p or q':^10} {'(not p or q) and p':^15}")
for p in valores:
    for q in valores:
        print(f"{str(p):^5} {str(q):^5} {str(not p or q):^10} {str((not p or q) and p):^15}")
```

```python
valores = [True, False]

print(f"{'p':^5} {'q':^5} {'r':^5} {'p or q':^10} {'not (p or q) or r':^20}")
for p in valores:
    for q in valores:
        for r in valores:
            print(f"{str(p):^5} {str(q):^5} {str(r):^5} {str(p or q):^10} {str(not(p or q) or r):^20}")
```
